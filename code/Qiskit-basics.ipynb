```python
from qiskit import *
```

```python
qc = QuantumCircuit()
qr = QuantumRegister(2)
qc.add_register(qr)
```

```python
qc.draw(output="mpl")
```

```python
qc.h(qr)
```

```python
qc.draw(output="mpl")
```

```python
qr2 = QuantumRegister(6)
```

```python
qc.add_register(qr2)
```

```python
qc.draw(output="mpl")
```

```python
qc.x(qr2)
```

```python
qc.draw(output="mpl")
```

```python
qc.cx(qr[1], qr2[5])
```

```python
qc.draw(output="mpl")
```

```python
Aer.backends()
```

```python
cr = ClassicalRegister(8)
qc.add_register(cr)
```

```python
qc.measure(qr[0], cr[0])
```

```python
backends = Aer.get_backend("qasm_simulator")
```

```python
job = execute(qc, backends)
```

```python
result = job.result()
```
